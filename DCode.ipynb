```python
# Split Train and Split
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:,df.columns != 'diabetes']
y = df.iloc[:,df.columns == 'diabetes']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
xtrain.head(3)
```

```python
xtest.head(3)
```

```python
ytest.head(3)
```

```python
ytrain.head(3)
```

```python
#Algorithm 1
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(xtrain,ytrain.values.ravel())
```

```python
predicted_output = model.predict(xtest)
print(predicted_output)
```

```python
from sklearn.metrics import accuracy_score
```

```python
ac = accuracy_score(ytest,predicted_output)
print(ac)
```

```python
# Algorithm 2
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model1 = KNeighborsClassifier()
```

```python
model1.fit(xtrain,ytrain.values.ravel())
```

```python
predict = model1.predict(xtest)
print(predict)
```

```python
acc1 = accuracy_score(predict,ytest)
print(acc1)
```

```python
# Finally i conclude that RandomForestClassifier produce higher accuracy in this comparison
```
